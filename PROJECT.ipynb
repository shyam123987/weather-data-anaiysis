<h2><b>Data cleaning</b></h2>

```python
#checking for the total null values
null_values = df.isnull().sum().sum()
print(f"There are {null_values} null_values")
```

```python
#DataFrame df by removing all rows (or columns) that contain missing values (NaN).
new_df = df.dropna()

print(new_df.to_string())
```

```python
# identity duplicate rows in data frame,Returns True for every row that is a duplicate, otherwise False
print(df.duplicated())
```

<h2><b>Data Visualization</b></h2>

```python
Temp_data = df["Temperature (Celsius)"]
ws_data = df["Wind Speed (m/s)"]

sns.kdeplot(Temp_data) 
plt.title("Temperature distribution")
plt.xlabel("temperature(celcius)")
plt.show()

sns.kdeplot(ws_data)
plt.title(" wind speed distribution")
plt.xlabel(" wind speed")
plt.show()
```

```python
condition_counts = df['Description'].value_counts()

plt.figure(figsize=(12,12))
plt.pie(condition_counts, labels=condition_counts.index, autopct='%1.1f%%', pctdistance=1.1, labeldistance=0.5, startangle=140)
plt.title('Distribution of Weather Conditions')
plt.axis('equal')
plt.show()
```

```python
temperature_by_city = df.groupby('City')['Temperature (Celsius)'].sum().sort_values()

plt.figure(figsize=(8,14))
count = temperature_by_city.plot(kind='barh', color='skyblue')
for i in count.containers:
    count.bar_label(i)
plt.title(' temperature by City')
plt.xlabel(' temperature (Celsius)')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
Temp_data = df["Temperature (Celsius)"]
ws_data = df["Wind Speed (m/s)"]

sns.scatterplot(x=Temp_data, y=ws_data, color='red')
plt.title("Temperature vs wind speed")
plt.xlabel("Temperature (°C)")
plt.ylabel("wind speed (m/s)")
plt.show()
```

```python
Temp_data = df["Temperature (Celsius)"]
city_data = df["City"]

plt.figure(figsize=(16,16))
count = plt.bar(Temp_data, city_data ,width=0.1,color = "#4CAF50")
#for i in count.containers:
plt.bar_label(count)

plt.title("City vs temperature")
plt.xlabel("Temperature (°C)",)
plt.ylabel("city")
plt.show()
```
